# Lenguaje python, crear la clase "animal" con los atributos "Nombre", "Raza" crear el constructor animal, y las funciones ladrar(), Correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definición de la clase
class Animal:
    # El constructor (__init__) inicializa los atributos del objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para ladrar
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau guau! 🐶")

    # Función para correr
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente. 🏃‍♂️💨")
    def Morder(self):
        print(f"{self.nombre} está mordiendo a la vecina")
# --- Creación de los objetos (Instancias) ---

# Objeto 1: Pitbull
pitbull = Animal("Zeus", "Pitbull")

# Objeto 2: Chihuahua
chihuahua = Animal("Chispa", "Chihuahua")
# Creamos el objeto  perro callejero con datos
Perrocallejero = Animal("solovino", "Chafa")

# --- Uso de los objetos ---

print(f"Perro 1: {pitbull.nombre}, Raza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()

print("-" * 30)

print(f"Perro 2: {chihuahua.nombre}, Raza: {chihuahua.raza}")
chihuahua.ladrar()
chihuahua.correr()

print("-" * 30)

print(f"Perro 3: {Perrocallejero.nombre}, Raza: {Perrocallejero.raza}")
Perrocallejero.ladrar()
Perrocallejero.correr()
print("-" * 30)
print("Programa realizado por Oscar Flores NL = 19")
```
